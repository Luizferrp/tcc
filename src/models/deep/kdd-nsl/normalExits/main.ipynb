```python
from lazylib import make, AlexNet, GoogLeNet, device
from lazylib.train import Epoch_anal
import torch
import json
```

```python
all_metrics, model = make(
  model=AlexNet(x=128).to(device()),
  epocs= 2, 
  csv_path = "../../../../../data/clean_nslkdd.csv",
  batch_size = 256
)
json.save(all_metrics, 'alexnet_nsl.json')
```

```python
metrics = all_metrics['treino'][-1]
Epoch_anal(metrics)
```
